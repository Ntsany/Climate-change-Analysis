#### #3 Analisis Rata-rata dan distribusi Suhu Global 
- Pertanyaan: Berapa rata-rata suhu global dalam dataset dan bagaimana persebarannya? 
- Tujuan: Memberikan ringkasan deskriptif dari suhu global untuk memahami karakteristik data
- Metode analisis: Statistik deskriptif seperti mean, median, standard deviation, dan distribusi suhu dengan histogram.

```python
# Statistik deskriptif suhu rata-rata global
country_temp_stats = data["Avg Temperature"].describe()

print(country_temp_stats)

# Mendapatkan Q1, Q3, dan IQR
Q1 = country_temp_stats["25%"]
Q3 = country_temp_stats["75%"]
IQR = Q3 - Q1

# Menghitung batas bawah dan batas atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menemukan data yang merupakan outlier
outliers = data[(data["Avg Temperature"] < lower_bound) | (data["Avg Temperature"] > upper_bound)]

# Menghitung jumlah outlier
outlier_count = len(outliers)

# Menampilkan jumlah outlier
print(f"Jumlah outlier: {outlier_count}")
```

Dari hasil analisis statistik terhadap rata-rata suhu global didapat insight:
- Nilai rata-rata suhu global adalah 19.88°C, dengan standar deviasi sebesar 8.54°C. Ini menunjukkan variasi suhu yang cukup besar di seluruh dunia.
- Suhu minimum adalah 5.00°C, sedangkan suhu maksimum mencapai 34.90°C, yang berarti cakupan rentang suhu cukup luas.
- Median suhu (20.10°C) yang mendekati rata-rata menunjukkan distribusi suhu yang cukup simetris tanpa dominasi data ekstrem.
- Tidak ditemukan adanya outlier berdasarkan IQR, sehingga distribusi data kemungkinan besar normal atau mendekati normal dan menunjukkan dataset tidak memiliki nilai-nilai ekstrem yang dapat memengaruhi analisis lebih lanjut.
- Rentang data dan distribusi kuartil memperlihatkan variasi suhu yang merepresentasikan keragaman iklim global, dari lokasi-lokasi bersuhu rendah hingga yang tinggi.


```python
# Plot histogram untuk distribusi suhu global
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x="Avg Temperature", bins=30, kde=True, color="skyblue")
plt.title("Distribusi Suhu Rata-Rata Global", fontsize=16)
plt.xlabel("Suhu Rata-Rata (°C)", fontsize=12)
plt.ylabel("Frekuensi", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Pada histogram diatas dapat dikatakan bahwa:
- Histogram menunjukkan distribusi suhu rata-rata global dengan rentang suhu sekitar 5°C hingga 35°C.
- Distribusi terlihat relatif merata, dengan sedikit puncak pada rentang 10°C - 15°C dan 25°C - 30°C.
- Peningkatan frekuensi suhu tertentu dapat menunjukkan wilayah dengan kondisi iklim spesifik (misalnya, subtropis yang lebih stabil).
- Garis lengkung di atas histogram menunjukkan kemungkinan distribusi normal, meskipun datanya memiliki penyebaran yang luas.

#### #6 Analisis Tren Kejadian Cuaca Ekstrem
- Pertanyaan: Apakah ada tahun tertentu yang menunjukkan lonjakan atau penurunan signifikan dalam jumlah kejadian cuaca ekstrem?
- Metode Analisis: Line graph  untuk memvisualisasikan kejadian cuaca ekstrem per tahun dengan analisis distribusi.

```python
# Filter data untuk rentang tahun yang tersedia (2000 hingga 2023)
data_recent_23_years = data[(data['Year'] >= 2000) & (data['Year'] <= 2023)]

# Agregasi jumlah kejadian cuaca ekstrem per tahun
extreme_events_per_year = data_recent_23_years.groupby('Year')['Extreme Events'].sum().reset_index()

# Membuat line plot untuk tren kejadian cuaca ekstrem
plt.figure(figsize=(12, 6))
sns.lineplot(data=extreme_events_per_year, x='Year', y='Extreme Events', marker='o', color='b')
plt.title("Tren Kejadian Cuaca Ekstrem (2000-2023)")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Kejadian Cuaca Ekstrem")

# Highlight dekade terakhir (2013 - 2023)
plt.axvspan(2013, 2023, color="orange", alpha=0.3, label="Dekade Terakhir")
plt.legend()

plt.show()
```

Dari gambar hasil visualisasi menunjukkan bahwa meskipun jumlah kejadian cuaca ekstrem secara keseluruhan menurun dari tahun 2000 ke 2010, dekade terakhir menunjukkan pola fluktuasi besar dengan pola naik-turun yang tidak stabil yang mungkin menjadi perhatian pembuat kebijakan.
- Lonjakan signifikan dari kejadian cauca ekstrim terjadi di tahun 2017 (di mana jumlah kejadian cuaca ekstrem meningkat tajam dibandingkan tahun sebelumnya) dan 2020 (jumlah kejadian cuaca ekstrem tertinggi dalam dekade terakhir)
- Penurunan signifikan cuaca ekstim: Terjadi pada 2016, ketika jumlah kejadian cuaca ekstrem menurun tajam dibandingkan tahun sebelumnya (2015).


#### #7 Analisis Variasi Tingkat Penggunaan Energi Terbarukan Antar Negara

- Pertanyaan: Bagaimana variasi penggunaan energi terbarukan? Negara mana yang memiliki penggunaan energi terbarukan paling tinggi dan rendah ditentukan berdasarkan variasi tersebut?
- Metode Visualisasi: Boxplot
- Langkah Analisis:
    - Hitung mean, standard deviation, dan variance dari kolom Renewable Energy.
    - Gunakan boxplot untuk menggambarkan sebaran data, termasuk outlier, median, dan kuartil.
    - Identifikasi negara-negara dengan penggunaan energi terbarukan yang tinggi dan rendah.

```python

# Analisis Statistik
mean_renewable = data['Renewable Energy'].mean()
std_renewable = data['Renewable Energy'].std()
var_renewable = data['Renewable Energy'].var()

print(f"Mean: {mean_renewable:.2f}")
print(f"Standard Deviation: {std_renewable:.2f}")
print(f"Variance: {var_renewable:.2f}")

# Visualisasi Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, y='Country', x='Renewable Energy', color='skyblue')
plt.title('Variasi Renewable Energy antar Negara', fontsize=14)
plt.ylabel('Country', fontsize=12)
plt.xlabel('Renewable Energy (%)', fontsize=12)
plt.show()
```

Berdasarkan boxplot ini, beberapa analisis dapat ditarik mengenai distribusi penggunaan energi terbarukan di berbagai negara:

1. Negara dengan Penggunaan Energi Terbarukan Tinggi:
    - Mexico memiliki median yang lebih tinggi, garis tengah pada kotak boxplot di 30%, menunjukkan bahwa separuh data penggunaannya ada di atas nilai tersebut. Dan mexico juga memiliki rentang antar kuartil yang luas.  
    - China juga memiliki median yang tinggi mencapai lebih dari 30%. Sehingga hal itu menunjukan bahwa China dan Mexico adalah negara dengan penggunaan energi terbarukan tertinggi.

2. Negara dengan Penggunaan Energi Terbarukan Rendah:
    - South Africa memiliki rentang kuartil yang kecil dibandingkan yang lain dan median yang paling rendah (sedikit diatas 20%). Sehingga bisa dikatakan bahwa dari data yang tercatat dalam dataset bahwa south africa termasuk negara yang sedikit menggunakan energi terbarukan.

3. Sebaran:
    - Negara seperti France dan Germany menunjukkan rentang yang cukup lebar, mengindikasikan adanya variasi yang signifikan di antara penggunaannya. Hal ini bisa saja dipengaruhi oleh kebijakan regional atau sumber daya lokal.
    - Sebaliknya, Japan memiliki rentang yang lebih sempit, mengindikasikan stabilitas dalam proporsi energi terbarukan di negara tersebut.
4. Negara dengan Outlier atau Anomali:

#### #8 Analisis Perubahan Forest Area dari Waktu ke Waktu
- Pertanyaan: Bagaimana perubahan kumulatif luas hutan dari waktu ke waktu?
- Metode Visualisasi: Area plot berguna untuk menampilkan akumulasi perubahan forest area sepanjang waktu. Ini akan memberi gambaran tentang bagaimana forest area berkembang dan memperlihatkan perubahan kumulatif secara lebih jelas.

#### #9 Proyeksi perubahan suhu rata-rata global dalam 10 tahun ke depan

- Pertanyaan: Berdasarkan tren saat ini, bagaimana proyeksi perubahan suhu rata-rata global dalam 10 tahun ke depan dengan tingkat akurasi minimal 85%?
- Metode Analisis: Analisis statistik inferensial menggunakan confidence interval atau regresi time series untuk proyeksi kenaikan suhu.
- Visualisasi: Line plot tren suhu rata-rata dan proyeksinya.



Proses:
- Data awal di proses dengan hitung rata-rata suhu glbal per tahun yang dikelompokkan berdasarkan tahun untuk menghasilkan tren tahunan dan kemudian dapat dipakai dalam analisis
- Tren suhu rata-rata divisualisasikan dengan line pllot untuk melihat pola historis bagaimana suhu berubah dari tahun ke tahun
- Regresi linier untuk proyeksi suhu: 
    - digunakan Year dan Avg Temperature sebagai variabel. 
    - Kemudian ditambahkan konstanta ke x sehingga model bisa mengambil intercept dalam regresi linier. 
    - lalu buat model regresi linier sederhana dengan Ordinary Least Squares dan tampilkan datanya agar mudah untuk dianalisis.
- Proyeksikan suhu 10 tahun ke depan: dimana tahun berikutnya dihitung berdasarkan tren saat ini. 
- Menghitung Confidence Interval (CI) untuk proyeksi dimana CI ini memperhitungkan ketidakpastian model dengan CI 95% untuk memberi rentang kemungkinan suhu yang lebih realistis.
- Lalu visualisasi proyeksi yang menggabungkan tren historis, proyeksi, dan ketidakpastian.
- Setelah itu, model di evaluasi dengan menghitung MAE (Mean Absolute Error) dan akurasi model.
    

```python
import statsmodels.api as sm
```

```python
# Menghitung rata-rata suhu global per tahun
global_avg_temp = data.groupby('Year')['Avg Temperature'].mean().reset_index()

# Visualisasi tren suhu
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Avg Temperature', data=global_avg_temp, marker='o')
plt.title('Tren Suhu Rata-rata Global')
plt.xlabel('Tahun')
plt.ylabel('Suhu Rata-rata (°C)')
plt.grid(True)
plt.show()

# Regresi Linier untuk proyeksi suhu global
X = global_avg_temp['Year'].values.reshape(-1, 1)  # Tahun sebagai variabel independen
y = global_avg_temp['Avg Temperature'].values  # Suhu sebagai variabel dependen

# Menambahkan konstanta untuk intercept (model linear)
X = sm.add_constant(X)

# Membangun model regresi
model = sm.OLS(y, X).fit()

# Menampilkan ringkasan model
print(model.summary())

# Proyeksi suhu untuk 10 tahun ke depan (tahun sekarang + 10 tahun)
future_years = np.array([global_avg_temp['Year'].max() + i for i in range(1, 11)]).reshape(-1, 1)
future_years_with_const = sm.add_constant(future_years)
future_temp_predictions = model.predict(future_years_with_const)

# Menampilkan proyeksi suhu
for year, temp in zip(future_years.flatten(), future_temp_predictions):
    print(f"Proyeksi suhu untuk tahun {year}: {temp:.2f}°C")

# Menghitung Confidence Interval (CI) untuk proyeksi
predictions_with_ci = model.get_prediction(future_years_with_const)
ci_lower, ci_upper = predictions_with_ci.conf_int(alpha=0.05).T

# Visualisasi proyeksi suhu dengan CI
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Avg Temperature', data=global_avg_temp, marker='o', label='Data Asli')
plt.fill_between(future_years.flatten(), ci_lower, ci_upper, color='gray', alpha=0.3, label='Confidence Interval (95%)')
plt.plot(future_years, future_temp_predictions, label='Proyeksi Suhu', color='red', linestyle='--')
plt.title('Proyeksi Suhu Rata-rata Global dengan Confidence Interval (95%)')
plt.xlabel('Tahun')
plt.ylabel('Suhu Rata-rata (°C)')
plt.legend()
plt.grid(True)
plt.show()

# Evaluasi model menggunakan MAE (Mean Absolute Error)
mae = np.mean(np.abs(y - model.predict(X)))
print(f"Mean Absolute Error (MAE) pada data training: {mae:.2f}")

# Cek akurasi model
accuracy = 100 - (mae / np.mean(y) * 100)
print(f"Akurasi model: {accuracy:.2f}%")

# Proyeksi akurasi minimal 85% berdasarkan MAE
if accuracy >= 85:
    print("Model mencapai akurasi minimal 85%.")
else:
    print("Model belum mencapai akurasi minimal 85%.")
```

Dapat dianalisis bahwa dari model OLS yang telah dibuat maka didapatkan:
- R-squared (0.025): Sangat rendah, menunjukkan bahwa model hanya menjelaskan 2.5% dari variabilitas data suhu rata-rata.
- Koefisien: Intercept (const): -36.3243 dan Slope (x1): 0.0279, menunjukkan tren kenaikan suhu sebesar 0.0279°C per tahun.
- Probabilitas F-statistic (0.464): Tidak signifikan, artinya hubungan antara tahun dan suhu rata-rata tidak cukup kuat untuk diandalkan secara statistik.
- Catatan Multikolinearitas: Kondisi besar (5.85e+05) menunjukkan kemungkinan masalah multikolinearitas atau numerik.

- Berdasarkan model regresi, suhu rata-rata global diproyeksikan meningkat secara linear, meskipun dengan tren yang sangat kecil (0.0279°C per tahun). Model memproyeksikan peningkatan suhu yang sangat lambat (hanya 0.0279°C per tahun) hingga tahun 2033.
- Tahun 2024 hingga 2033 menunjukkan kenaikan suhu bertahap dari 20.15°C menjadi 20.40°C
- MAE sebesar 0.95°C menunjukkan rata-rata kesalahan prediksi model pada data historis.
- Akurasi model mencapai 95.20%, jauh di atas syarat minimal 85%. Meskipun akurasi model tinggi (95.20%), rendahnya R-squared (0.025) menunjukkan bahwa model ini tidak menangkap banyak variabilitas dalam data.
